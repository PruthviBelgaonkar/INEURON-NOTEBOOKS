```python
import pandas as pd
df=pd.read_csv(r"D:\New folder\income.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_pred=km.fit_predict(df[['Age','Income($)']])
```

```python
y_pred
```

```python
df['cluster']=y_pred
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_pred=km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,-1])
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
km.cluster_centers_
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
```
