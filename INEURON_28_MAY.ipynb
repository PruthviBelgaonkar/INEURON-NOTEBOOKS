```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("E:/ineuronbasics/airfoil_self_noise.dat",sep='\t',header=None)
```

```python
df
```

```python
df.columns=['Frequency','Angle of Atttack','Chord length','Free-stream velocity','Suction Side','Pressure Level']
```

```python
df
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```
